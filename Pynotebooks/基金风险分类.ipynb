```python
import pandas as pd
```

```python
df = pd.read_excel('E:/附件二：场外开放式基金维度信息0615.xlsx','拟调整后评级')
dfmap0 = pd.read_excel('e:/map.xlsx','保本')
dfmap1 = pd.read_excel('e:/map.xlsx','债券')
bblist = [x[1:] for x in dfmap0['代码']]
zzlist = [x[1:] for x in dfmap1['代码']]
```

```python
def my_map_func(x):
    if x['风险等级']=='低' and x['投资品种']=='固定收益类' and x['基金代码'][:6] in zzlist:
        x['风险等级'] = '中低'
        print(x['风险等级'],x['基金名称'])
    elif x['风险等级']=='中'and x['投资品种']=='权益类' and x['基金代码'][:6] not in bblist:
        x['风险等级'] = '中高'
        print(x['风险等级'],x['基金名称'])
    return x
```

```python
df1 = pd.DataFrame(index=df.index,columns=df.columns)
for x in range(len(df.index)):
    df1.iloc[x,:] = my_map_func(df.iloc[x,:])
```

```python
df1.to_excel('e:/adjusted.xlsx','19日调整后评级')
```
